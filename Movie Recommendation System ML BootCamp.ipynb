```python
movies_title=pd.read_csv('u.item',sep="\|",header=None,encoding="ISO-8859-1")
```

```python
movies_titles=movies_title[[0,1]] #Extracting move id and title (position 0 and 1) from movies_title and saving in movies_titles
movies_titles.columns=["item_id","title"] #Naming columns
movies_titles.head()
```

```python
df=pd.merge(df,movies_titles,on="item_id") #Merging 2 data sets - df and movies_titles on common column item id
```

```python
ratings=pd.DataFrame(df.groupby('title').mean()['rating']) #Saving titles and avrage rating of that movie in ratings 
```

```python
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title').count()['rating']) #counting the number of peoples rating a movie
ratings.head()
```

#Create the recommendar System

```python
df.head()
```

```python
moviemat=df.pivot_table(index="user_id",columns="title",values="rating")
```

```python
moviemat.head()
```

```python
starwars_user_ratings=moviemat['Star Wars (1977)'] #Extracting all users rating for star wars movie
```

```python
starwars_user_ratings.head(20)
```

```python
similar_to_starwars=moviemat.corrwith(starwars_user_ratings) #Extracting correlated movies for star wars from moviemat
```

```python
similar_to_starwars
```

```python
#Creating a table with movies similar to star wars and there correlation with it.
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['Correlation']) 
```

```python
corr_starwars.dropna(inplace=True) #Removing NaN values
```

```python
corr_starwars
```

```python
corr_starwars.head()
```

```python
#Sorting correlation in decending order to find perfect correlated movie to star wars with value 1.0(positive)
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars=corr_starwars.join(ratings['num of ratings']) #Joining 2 tables(corr_starwars & ratings column(number of ratings))
#Merge can also be used
```

```python
corr_starwars.head()
```

```python
#Extracting movies with 100+ number of ratings and sorting Correlation in desending order.
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False)
```

```python
def predict_movies(movie_name):
    movie_user_ratings=moviemat[movie_name]
    similar_to_movie=moviemat.corrwith(movie_user_ratings)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['Correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie=corr_movie.join(ratings['num of ratings'])
    
    predictions=corr_movie[corr_movie['num of ratings']>100].sort_values('Correlation',ascending=False)
    
    return predictions
```

```python
predict_my_movie=predict_movies("Titanic (1997)")
```
